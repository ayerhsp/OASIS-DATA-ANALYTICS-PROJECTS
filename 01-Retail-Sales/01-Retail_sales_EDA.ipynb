```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('store_sales.csv')

print("Shape:", df.shape)
print("Missing values:")
print(df.isnull().sum())

print("\nBasic Stats:")
df.describe()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10,5))
sns.countplot(data=df, x='Category')
plt.xticks(rotation=45)
plt.title("Sales Count by Category")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.barplot(data=df, x='Gender', y='Amount')
plt.title("Average Spending: Male vs Female")
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.boxplot(data=df, x='Season', y='Amount')
plt.title("Spending Distribution by Season")
plt.show()
```

```python
top_items = df['ItemPurchased'].value_counts().head(10)
plt.figure(figsize=(10,5))
sns.barplot(x=top_items.values, y=top_items.index)
plt.title("Top 10 Most Purchased Items")
plt.show()
```

```python
plt.figure(figsize=(10,6))
numeric_df = df.select_dtypes(include=['int64','float64'])
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Between Variables")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.countplot(data=df, x='PaymentMethod')
plt.title("Most Used Payment Method")
plt.show()
```
