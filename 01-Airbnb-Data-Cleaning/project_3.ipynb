```python
!pip install -q kagglehub
import kagglehub
import pandas as pd
path = kagglehub.dataset_download("dgomonov/new-york-city-airbnb-open-data")
df = pd.read_csv(f'{path}/AB_NYC_2019.csv')
print("Original:", df.shape)
df['name'] = df['name'].fillna('Unknown')
df['host_name'] = df['host_name'].fillna('Unknown')
df['last_review'] = pd.to_datetime(df['last_review'])
df['reviews_per_month'] = df['reviews_per_month'].fillna(0)
df = df.drop_duplicates()
df = df[df['price'] < 1000]
df = df[df['price'] > 0]
df['neighbourhood_group'] = df['neighbourhood_group'].str.upper()
df['room_type'] = df['room_type'].str.title()
print("CHECK 1 - Nulls:", df.isnull().sum().sum())
print("CHECK 2 - Shape after dups:", df.shape)
print("CHECK 3 - Min Price:", df['price'].min())
print("CHECK 4 - Max Price:", df['price'].max())
df.to_csv('AB_NYC_2019_Cleaned.csv', index=False)
print("FINAL Cleaned Shape:", df.shape)
df.head()
```
